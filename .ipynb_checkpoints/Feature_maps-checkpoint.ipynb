# Displaying ConvNets feature maps on real-time video with Keras and OpenCV

In this project, we used Python and OpenCV to display feature maps of a ConvNet trained for an mini RC autonomous car. The full article is avaliable on [Medium](https://keras.io).

The following techniques are used:
- Feature maps extraction
- Image binarisation
- Image segmentation
- Bitwise operations

Finally, a .AVI video is created merging input images and relevant neurons activations.

## Necessary imports
The following libraries are used: <br/>
        `tqdm` keep track of processing time on iterators<br/>
        `cv2` image processing with OpenCV <br/>
        `Keras` handling Keras models <br/>
        `functions` contains a set of self-made useful functions


```python
import warnings
warnings.filterwarnings("ignore")

from keras.models import load_model
from keras.utils.vis_utils import plot_model


import matplotlib.pyplot as plt
from pylab import rcParams
from tqdm import tqdm


from functions import *
import cv2 
```

## Load Keras autopilot ConvNet

```python
model_path = './Models/'

# Loading Keras model
model = load_model(model_path + 'autopilot.hdf5')

# Print model summary
model.summary()
```

## Load images

```python
# Loading images from the provided database
img_path = './Datasets/high_res_images/'
imgs = load_images(img_path)

print('Image shape = ', np.shape(imgs[0]))
```

## Display one example of input images

```python
# Set Matplotlib wide image size
rcParams['figure.figsize'] = 16, 40

# Display images on RGB color
plt.imshow(cv2.cvtColor(imgs[0], cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Original image')
```

## Processing functions

```python
def pre_process(img):
    """ Process image to be feed into a Keras convolutive model.
    
    img: 3-channel Numpy array.
    Returns a resized and smoothed (250, 70, 3) image. 
    """
    # Crop relevant area from input image
    img = crop(img,600,1080,0,1800)
    # Resize to model's input size (250,70,3)
    img = cv2.resize(img,(250,70), interpolation = cv2.INTER_LANCZOS4)#INTER_CUBIC)
    # Smooth image with mean filter
    kernel = np.ones((2,2),np.float32) / 4
    img = cv2.filter2D(img,-1,kernel)

    return img


def merge(target_map, cropped_img, min_thres=80):
    """ Merge real input image and feature maps activations.
    
    targete_map: 3-channel Numpy array target map.
    cropped_img: 3-channel Numpy array with input image ROI.
    min_thres: lower threshold on image binarisation
    """

    # Get numpy array from matplotlib canvas
    target_map = get_figure(target_map, cmap='inferno').astype('uint8')
    # Crop unnecessary regions
    target_map = crop(target_map,9,63,33,232)
    # Resize to original size
    target_map = cv2.resize(target_map,(1800,480), interpolation = cv2.INTER_CUBIC)

    # Create binary mask
    img2gray = cv2.cvtColor(target_map,cv2.COLOR_BGR2GRAY)
    ret, mask = cv2.threshold(img2gray, min_thres, 255, cv2.THRESH_BINARY)
    # Apply mask to activation map
    dst_map = cv2.bitwise_and(target_map, target_map, mask=mask)

    # Convert RGB to BGR (optional)
    #dst_map = cv2.cvtColor(dst_map, cv2.COLOR_RGB2BGR)
    # Add images
    final_img = cv2.add(dst_map, cropped_img)

    return final_img
```

## Generate merged images

```python
# Generate merged images
merged_maps = []

# Get desired feature map (2nd convolutive layer)
layer_name = model.layers[6].name

for i in tqdm(range(len(imgs))):

    # Save input image ROI 
    cropped_img = crop(imgs[i],600,1080,0,1800)
    
    # Transform original image to model's input size (250, 70, 3)
    dst = pre_process(imgs[i])
    
    # Input image to Keras model. Don't forget the mini-batch dimension
    activations = get_activations(model,np.expand_dims(dst, axis=0), 
                                  print_shape_only=True, 
                                  layer_name=layer_name)
    
    # Normalize target map
    target_map = norm_pixels(activations[0][0][:,:,1])

    # Create merged image. You may want to change the lower threshold.
    dst = merge(target_map, cropped_img, 80)
    
    # Append image to list
    merged_maps.append(dst)
```

## Sequence preview

```python
# Display a preview sequence before writing video
plot_sequence(merged_maps,0,15)
```

## Write .AVI video

```python
# Create video writer
writer = cv2.VideoWriter("sample_output.avi", 
                         cv2.VideoWriter_fourcc(*"MJPG"), 
                         30,(1800,480))

for frame in tqdm(merged_maps):
    writer.write(frame.astype('uint8'))
    
writer.release()
```
